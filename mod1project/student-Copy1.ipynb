```python
#input first wave of dependencies
import pandas as pd #for multidimensional arrays like DataFrames and Series
import numpy as np #for easy calculations of numerical data
import statistics
```

```python
#import the excel from data.world
orig_dataset = pd.read_excel('https://query.data.world/s/gbgavrlgern5guuyq5yd5tkp4amzl3')    
```

```python
orig_dataset.drop('MovieID', axis=1, inplace=True)
#remove the MovieID column (unnecessary)
```

```python
orig_dataset.head()
#show the top 5 values
```

```python
df = orig_dataset.dropna().sort_values(by = ['Release Date'], ascending=False).head(250)
#create a new dataframe that drops every NaN and NaT value, and sorts by release date (latest to oldest movie), and includes 100 movies
```

```python
df.reset_index(inplace=True, drop=True)
#create a new index
```

```python
df
#lets see what we get
```

```python
#Need to create a Profit column using Gross and Budget and also create a day of the week column
df['Profit'] = df['Gross'] - df['Budget']
df['Day of Week'] = df['Release Date'].dt.day_name()
df['Year'] = df['Release Date'].dt.year
#since im already here, I also want to create quaters of the year
df['Quarter'] = df['Release Date'].dt.quarter 
```

```python
#Let's see out updated DataFrame
df.head()
```

```python
#I found an outlier with a super high profit when I was sorting movies by profit and weekday
#This movie was on a Thursday

df[df['Day of Week']=='Thursday']
```

```python
#Delete anything above 1 billion for profit
df = df[df['Profit'] <= 1000000000]
#The movie specifically was Avatar
```

```python
print(df.dtypes)
```

```python
#now we use scipy stats
arr = np.array(df['Profit'])
arr.tolist()
print(arr)
```

```python
profit_mean = statistics.mean(arr)
print(profit_mean)
```

```python
#import seaborn and other dependencies that may be needed for time and plotting

import seaborn as sns
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set(style='white')
sns.set_context("paper", font_scale=1.5) 
f, ax = plt.subplots(figsize=(15, 5))
sns.countplot(y="MPAA Rating", data=df)
sns.despine();
#DONE
```

```python
f, ax = plt.subplots(figsize=(12, 8))

# Plot the Gross
sns.set_color_codes("pastel")
sns.set_style('ticks')
sns.barplot(x="Gross", y="Genre", data=df,
            label="Gross", color="b", ci=None)

# Plot the Budget
sns.set_color_codes("muted")
sns.barplot(x="Profit", y="Genre", data=df,
            label="Profit", color="b", ci=None)

#Plot a vertical 'Profit' mean, with a label on the line
 
ax.axvline(int(profit_mean), color='red',linestyle='dashed', linewidth=2.0)
plt.text(int(profit_mean), 7, 'avg. 3.7mil profit',fontsize='15', color='red')

# Add a legend and informative axis label
ax.legend(ncol=2, loc="upper right", frameon=True, fontsize='13')
ax.set(xlim=int(), ylabel="",
       xlabel="USD (million)")
ax.get_xaxis().get_major_formatter().set_scientific(False)
plt.xscale('linear')
sns.despine(left=True, bottom=True)
```

```python
sns.set(style='white')
f, ax = plt.subplots(figsize=(14, 10))
# Draw a categorical scatterplot to show each observation
sns.swarmplot(x="Quarter", y='Year', hue="Genre",
              data=df)
sns.despine(offset=10, trim=True)
ax.legend(ncol=1, loc="center", frameon=True, fontsize='10');
```

```python
f, ax = plt.subplots(figsize=(14, 10))
sns.set_style('white')
sns.lineplot(data=df, x='Year', y='Rating',
             hue='MPAA Rating',
             err_style=None
            )
ax.set_xlim([2000, 2016])
ax.set_ylim([4, 10])
sns.despine(offset=10, trim=True);
```
